```python
import pandas as pd
import numpy as np
import IPython.core.display as display #reference=> https://gist.github.com/ksomemo/d3d1a47baebcceeafef949895612ec92
import IPython.display 

N=16

nl=[]
summn=[]
for m in range(0,N):
    summ=[]
    nl.append(m)
    for n in range(0,N):
        r=(m+n)%N
        q=(m+n-r)//N
        summ.append((q,r))
        
    summn.append(summ)

prodmn=[]
for m in range(0,N):
    prodm=[]
    for n in range(0,N):
        r=(m*n)%N
        q=(m*n-r)//N
        prodm.append((q,r))
        
    prodmn.append(prodm)    

df=[pd.DataFrame(summn),pd.DataFrame(prodmn)]
IPython.display.display(df[0])
print("index: +")
IPython.display.display(df[1])
print("index: ×")
```
